```python
import numpy as np
import cv2
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
img_ndarray = cv2.imread('samples/l4_160_3104_1_1_000000446_transf.jpg')
imshow(img_ndarray)
```

```python
f = open('samples/l4_160_3104_1_1_000000446_transf.txt','r')
bbox_pts = f.readline()
f.close()
print(bbox_pts)
```

```python
h,w = img_ndarray.shape[0:2]
print('w: %s | h : %s ' % (w, h))
```

```python
bbox_pts = bbox_pts.replace('\n','')
pontos = bbox_pts.split(',')[0:8]
print(pontos)
pontos = [int(float(ponto)*w) if indice < 4 else int(float(ponto)*h)  for indice, ponto in enumerate(pontos)]
print(pontos)
array_pts = [[pontos[i],pontos[4+i]] for i in range(4)]
print(array_pts)
```

```python
pts = np.array(array_pts).reshape((-1, 1, 2))
color = (255, 0, 0)
# Line thickness of 2 px
thickness = 2
# Using cv2.polylines() method
# Draw a Blue polygon with 
# thickness of 1 px
isClosed = True
image = cv2.polylines(img_ndarray, [pts], isClosed, color, thickness)
imshow(image)
```
